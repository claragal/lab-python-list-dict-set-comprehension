````python
def initialize_inventory(products):
    inventory = {product: int(input(f"Introduce the quantity for '{product}': ")) for product in products}
    return inventory

def get_customer_orders(products):
    try:
        num_orders = int(input("Enter the number of customer orders: "))
        if num_orders < 0:
            raise ValueError("Number of orders cannot be negative.")
    except ValueError as e:
        print(f"Invalid input: {e}. Please enter a valid number.")
        return set()

    customer_orders = {input(f"Enter the name of product {i+1}: ") for i in range(num_orders) if input(f"Enter the name of product {i+1}: ") in products}
    return customer_orders

def update_inventory(inventory, customer_orders):
    inventory = {product: (quantity - 1 if product in customer_orders and quantity > 0 else quantity) 
                 for product, quantity in inventory.items()}
    inventory = {product: quantity for product, quantity in inventory.items() if quantity > 0}
    for product in customer_orders:
        if product not in inventory:
            print(f"Sorry, {product} is out of stock.")
    return inventory

def calculate_order_statistics(customer_orders, products):
    total_products_ordered = len(customer_orders)
    percentage_products_ordered = (total_products_ordered / len(products)) * 100
    return total_products_ordered, percentage_products_ordered

def print_order_statistics(total_products_ordered, percentage_products_ordered):
    print("Order statistics:")
    print(f"Total products ordered: {total_products_ordered}")
    print(f"Percentage of Products Ordered: {percentage_products_ordered:.2f}%")

def print_updated_inventory(inventory):
    print("\nUpdated inventory:")
    [print(f"{product}: {quantity}") for product, quantity in inventory.items()]

def calculate_total_price(customer_orders):
    prices = {product: float(input(f"Enter the price for '{product}': ")) for product in customer_orders}
    total_price = sum(prices.values())
    return total_price

def main():
    products = ["t-shirt", "mug", "hat", "book", "keychain"]
    print("Products:", products)

    inventory = initialize_inventory(products)
    print("Inventory:", inventory)

    customer_orders = get_customer_orders(products)
    print("Customer Orders:", customer_orders)

    total_products_ordered, percentage_products_ordered = calculate_order_statistics(customer_orders, products)
    print_order_statistics(total_products_ordered, percentage_products_ordered)

    total_price = calculate_total_price(customer_orders)
    print(f"Total price of the order: {total_price:.2f}")

    inventory = update_inventory(inventory, customer_orders)
    print_updated_inventory(inventory)

if __name__ == "__main__":
    main()
````
